```python
import os
import chromadb
from dotenv import load_dotenv
from openai import OpenAI
from langchain_openai import OpenAIEmbeddings
```

```python
load_dotenv()

client = OpenAI()
embedder = OpenAIEmbeddings(model="text-embedding-3-small")
chroma_client = chromadb.PersistentClient(path="./data")
collection = chroma_client.get_or_create_collection(name="sessions")
```

```python
def get_similar_chunks(query, collection, top_k=5):
    query_embedding = embedder.embed_query(query)
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )

    # documents = []
    # for i, doc in enumerate(results['documents'][0]):
    #     metadata = results['metadatas'][0][i]
    #     metadata['abstract'] = doc
    #     documents.append(metadata)
    
    return results['documents'][0]
```

```python
def generate_answer(query, similar_chunks):
    context = ""
    for chunk in similar_chunks:
        context += f"Title: {chunk['title']}\nSpeakers: {chunk['speakers']}\nAbstract: {chunk['abstract']}\n\n"

    messages = [
        {"role": "system", "content": "You are a helpful assistant helping attendees at Nebraska.Code(), a software development conference, make the most of their conference by answering questions about sessions and helping them find the best sessions to attend based on their interests. Your answers should be concise, but conversational - do not simply respond with the verbatim session data, but include relevant information such as the room and times of the sessions."},
        {"role": "user", "content": f"Question: {query}\n\nSessions related to the user's question:\n{context}"}
    ]

    print(context)

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        max_tokens=500
    )

    return response.choices[0].message.content
```

```python
query = "I really like Adam Barney. Is he giving any sessions this year?"
similar_chunks = get_similar_chunks(query, collection)
answer = generate_answer(query, similar_chunks)

print("Answer:", answer)
```
